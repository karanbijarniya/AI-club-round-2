```python
import tensorflow as tf
from tensorflow.keras import layers, models, optimizers

def build_ser_cnn(input_shape, num_classes):
    """
    Builds the baseline CNN architecture as described in Section 6.3.
    Architecture:
    - 2 Convolutional Layers
    - Max/Average Pooling
    - 1 Fully Connected Layer (1024 neurons)
    - Softmax Output (5-way or 7-way)
    """
    model = models.Sequential()

    # Layer 1: Convolutional (8 filters, 5x5 kernel, stride 1)
    model.add(layers.Conv2D(8, (5, 5), strides=(1, 1), padding='same',
                            activation='relu', input_shape=input_shape))
    model.add(layers.MaxPooling2D((2, 2), strides=(2, 2))) # [cite: 201, 203, 204]

    # Layer 2: Convolutional (16 filters, 5x5 kernel)
    model.add(layers.Conv2D(16, (5, 5), padding='same', activation='relu'))
    model.add(layers.MaxPooling2D((2, 2), strides=(2, 2))) # [cite: 203, 204]

    # Flatten for Dense Layer
    model.add(layers.Flatten())

    # Fully Connected Layer (1024 hidden neurons)
    model.add(layers.Dense(1024, activation='relu'))
    model.add(layers.Dropout(0.5)) # Dropout used to reduce overfitting [cite: 199, 208]

    # Output Layer (Softmax unit)
    model.add(layers.Dense(num_classes, activation='softmax')) # [cite: 200]

    # Optimizer: Adam | Loss: Cross-entropy
    optimizer = optimizers.Adam()
    model.compile(optimizer=optimizer,
                  loss='categorical_crossentropy',
                  metrics=['accuracy']) #

    return model

# Example initialization for a 129x129 spectrogram [cite: 214]
# num_classes would be 5 or 7 depending on the database [cite: 200]
model = build_ser_cnn(input_shape=(129, 129, 1), num_classes=5)
model.summary()
```
