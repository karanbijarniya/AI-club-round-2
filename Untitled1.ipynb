```python
import numpy as np

def add_white_gaussian_noise(signal, snr_db=15):
    """
    Adds Gaussian noise to audio signal based on +15 SNR[cite: 185].
    """
    # Calculate average power of the signal
    p_signal = np.mean(signal**2)

    # Calculate noise power based on SNR formula [cite: 186]
    # SNR = 10 * log10(P_signal / P_noise)
    p_noise = p_signal / (10**(snr_db / 10))

    # Generate white Gaussian noise
    noise = np.random.normal(0, np.sqrt(p_noise), signal.shape)

    return signal + noise
```

```python
def jaccard_similarity(paper_x, paper_y):
    """
    Calculates Jaccard Similarity between two binary attribute vectors[cite: 136, 140].
    Formula: Z11 / (Z01 + Z10 + Z11)
    """
    x = np.array(paper_x)
    y = np.array(paper_y)

    # Z11: Both have value 1 [cite: 141]
    z11 = np.sum((x == 1) & (y == 1))

    # Z01: X is 0, Y is 1 [cite: 142]
    z01 = np.sum((x == 0) & (y == 1))

    # Z10: X is 1, Y is 0 [cite: 143]
    z10 = np.sum((x == 1) & (y == 0))

    denominator = z01 + z10 + z11

    if denominator == 0:
        return 0.0

    return z11 / denominator # [cite: 139]
```
